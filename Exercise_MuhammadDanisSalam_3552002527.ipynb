```python
#exercise pertama
#program untuk mengecek boleh tidaknya membuat SIM
#sesuai dengan umur yang disyaratkan

umur = int (input('Tolong masukkan umur-mu = '))

if umur >= 17:
  print ('kamu boleh membuat SIM')
else:
  print ('kamu tidak boleh membuat SIM')
```

```python
#Exercise kedua
#program untuk menentukan nilai yang dimasukan

nilai = int(input('masukan nilai-mu = '))

if nilai >= 90:
  print('Grade A')
elif nilai >= 80:
  print('Grade B+')
elif nilai >= 70:
  print('Grade B')
elif nilai >= 60:
  print('Grade C+')
elif nilai >= 50:
  print('Grade C')
elif nilai >= 40:
  print('Grade D')
else:
  print('Grade E')
```

```python
#Exercise ketiga
#program untuk menentukan grade nilai yang dimasukan

nilai = float(input('Masukan nilai mu ='))

if (nilai >= 80.00) and (nilai <= 100.00):
  print('Grade A')
elif (nilai >= 77.50) and (nilai <= 79.99):
  print('Grade A-')
elif (nilai >= 75.00) and (nilai <= 77.49):
  print('Grade A/B')
elif (nilai >= 72.50) and (nilai <= 74.99):
  print('Grade B+')
elif (nilai >= 70.00) and (nilai <= 72.49):
  print('Grade B')
elif (nilai >= 67.50) and (nilai <= 69.99):
  print('Grade B-')
elif (nilai >= 65.00) and (nilai <= 67.49):
  print('Grade B/C')
elif (nilai >= 62.50) and (nilai <= 64.99):
  print('Grade C+')
elif (nilai >= 60.00) and (nilai <= 62.49):
  print('Grade C')
elif (nilai >= 55.00) and (nilai <= 59.99):
  print('Grade C-')
elif (nilai >= 50.00) and (nilai <= 54.99):
  print('Grade C/D')
elif (nilai >= 45.00) and (nilai <= 49.99):
  print('Grade D+')
elif (nilai >= 40.00) and (nilai <= 49.99):
  print('Grade D')
else:
  print('Grade E')
```
