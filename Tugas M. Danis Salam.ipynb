Latihan 1

```python

# NPM utama
npm_saya = 3552002527

# 4 NPM lain (sebelum / setelah, sesuai data yang kamu berikan)
npm_lain = [3552002511, 3552002529, 3552002567, 3552002535]

# Membuat array berisi NPM saya + 4 NPM lainnya
npm_array = [npm_saya] + npm_lain

# Menampilkan isi array
print("Isi Array NPM:")
for npm in npm_array:
    print(npm)
```

```python

# Operasi Matematika Sederhana

angka = [2500, 5000, 7500, 10000, 20000]

print("Harga total:", sum(angka))
print("Harga dengan nilai tertinggi:", max(angka))
print("Harga dengan nilai termurah:", min(angka))
```
